```python
#Importing Necessary Libraries
import pandas as pd
import tabula
import PyPDF2
import matplotlib.pyplot as plt
```

```python
#Loading PDF Dataset from the Directory

pdf_file = 'Data/APEX_Loans_Database_Table (5).pdf'
with open(pdf_file, 'rb') as file:
    pdf_reader = PyPDF2.PdfReader(file)
    df_pdf = pd.DataFrame([page.extract_text().split('\n') for page in pdf_reader.pages])
df_pdf.head()
```

```python
# Loading data from Excel Directory

excel_path = 'Data/APEX Loan Data (5).xlsx'
df_excel = pd.read_excel(excel_path)
df_excel.head()
```

```python
# Checking for missing values in the PDF Dataset

print("Missing values in PDF data:")
print(df_pdf.isnull().sum())
```

```python
#Checking for Missing Values in the Excel Dataset

print("Missing values in Excel data:")
print(df_excel.isnull().sum())
```

```python
# Filling & Dropping missing values in the two datasets before Merging them accordingly
df_pdf.fillna(method='ffill', inplace=True)
df_excel.dropna(inplace=True)
```

```python
# Combining the two datasets
merged_data = pd.concat([df_excel, df_pdf], ignore_index=True)

merged_data.head()
```

```python
# Checking the data types
print(merged_data.dtypes)
```

```python
# Checking for duplicates
print(f"Number of duplicates: {merged_data.duplicated().sum()}")

# Handling missing values
combined_data = merged_data.fillna(0)
```

```python
#Printing the Merged Dataset Columns and Rows
merged_data.shape
```

```python
merged_data.describe()
```

<h2 style="color:brown">Descriptive Analysis</h2>

```python
# Calculating the Total Amount Loaned
total_amount_loaned = merged_data['LoanAmount'].sum()
print(f"Total Amount Loaned: {total_amount_loaned}")
```

```python
# Calculating the Average Amount Loaned
average_amount_loaned = merged_data['LoanAmount'].mean()
print(f"Average Amount Loaned: {average_amount_loaned}")
```

```python
# Calculating the Average Loan Term
print(f"Average loan term: {merged_data['Loan_Amount_Term'].mean()}")
```

```python
# Applicant Breakdown by Gender
gender_breakdown = merged_data.groupby('Gender')['Loan_Status'].value_counts(normalize=True).unstack()
print("Applicant Breakdown by Gender:")
print(gender_breakdown)
```

```python
# Generating visualization for the breakdown of approved and rejected loans by gender
plt.figure(figsize=(8, 6))
merged_data.groupby(['Loan_Status', 'Gender'])['Loan_ID'].count().unstack().plot(kind='bar')
plt.xlabel('Loan Status')
plt.ylabel('Number of Loans')
plt.title('Loan Approval by Gender')
plt.show()
```

```python
# Calculating the Maximum and minimum loan amounts
max_loan_amount = merged_data['LoanAmount'].max()
min_loan_amount = merged_data['LoanAmount'].min()
print(f"Max Loan Amount: {max_loan_amount}")
print(f"Min Loan Amount: {min_loan_amount}")
```

```python
# Calculating the percentage of self-employed applicants who had their loan approved
self_employed_approved = merged_data[(merged_data['Loan_Status'] == 'Y') & (merged_data['Self_Employed'] == 1)].shape[0]
total_approved = merged_data[merged_data['Loan_Status'] == 'Y'].shape[0]
self_employed_approval_percentage = (self_employed_approved / total_approved) * 100
print(f"Percentage of self-employed applicants who had their loan approved: {self_employed_approval_percentage:.2f}%")
```

```python
# Visualizing the percentage of self-employed applicants who had their loan approved
plt.figure(figsize=(8, 6))
plt.pie([self_employed_approved, total_approved - self_employed_approved], labels=['Self-employed Approved', 'Others Approved'], autopct='%1.1f%%')
plt.title('Loan Approval by Self-Employment Status')
plt.show()
```

```python
# Calculating the percentage of self-employed who had their loan approved from the merged dataset
self_employed_approval = merged_data.groupby(['Self_Employed', 'Loan_Status']).size().unstack()
self_employed_approval_percentage = self_employed_approval.div(self_employed_approval.sum(axis=1), axis=0) * 100

# Plotting the percentage of self-employed applicants with loan approval
ax = self_employed_approval_percentage.plot(kind='bar', stacked=False)
plt.title('Percentage of Self-Employed Applicants with Loan Approval')
plt.xlabel('Self_Employed')
plt.ylabel('Percentage')

# Function to add percentage annotations on top of each bar
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.2f}%', (x + width/2, y + height), ha='center', va='center', color='black', fontsize=10)

plt.legend(title='Loan_Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Income distribution of all main applicants
print(f"Average applicant income: {merged_data['ApplicantIncome'].mean()}")
print(f"Standard deviation of applicant income: {merged_data['ApplicantIncome'].std()}")

plt.figure(figsize=(8, 6))
merged_data['ApplicantIncome'].hist(bins=20)
plt.xlabel('Applicant Income')
plt.ylabel('Frequency')
plt.title('Distribution of Applicant Income')
plt.show()
```

```python
# Identifying the Top ten applicants by loan amount
top_ten_applicants = merged_data.nlargest(10, 'LoanAmount')[['Loan_ID', 'ApplicantIncome', 'LoanAmount']]
print("\nTop Ten Applicants by Loan Amount:")
print(top_ten_applicants)
```

```python
#Distribution of properties
property_distribution = merged_data.groupby('Property_Area')['Loan_ID'].count()
print("Distribution of properties:")
print(property_distribution)
```

```python
# Distribution of properties Visualization
property_distribution = merged_data['Property_Area'].value_counts()
property_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Properties for Loan Applicants')
plt.show()
```
